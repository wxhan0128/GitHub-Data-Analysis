```python
# Import data

import csv
with open('gh_lan_pca.csv') as f:
    next(f) # Skip header
    data = [tuple(map(float, row)) for row in csv.reader(f)]
    
n=28
matrix = [[0] * n for i in range(n)]

for user, item, rating in data:
    matrix[int(user)-1][int(item)-1] = rating # Convert to 0-based index.
    
# Processing data
big_list = []
for row in matrix:
    small_list = []
    for item in row:
        small_list.append(item)
    small_array = np.asarray(small_list)
    big_list.append(small_array)
big_array = np.asarray(big_list)
X = big_array

#X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])  
outFile2 = open("g4.txt", "w")

pca=PCA()
pca.fit(X)

ev=pca.explained_variance_
ev_ratio=[]

for i in range(len(ev)):
    ev_ratio.append(ev[i]/ev[0])
    
n=0
for i in range(len(ev_ratio)):
    if ev_ratio[i]<0.05:
        n=i
        break
    

pca = PCA(n_components=n)  
principalComponents = pca.fit_transform(X)
print (principalComponents)
for item in principalComponents:
    lines=(str(str(item) + "\n"))
    outFile2.write(lines)
x=[]
y=[]
for item in principalComponents:
    #if item[1]<4 and item[1]>-4 and item[0]<4:
    x.append(item[0])
    y.append(item[1])
plt.plot(x,y, "o")
```
